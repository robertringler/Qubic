# Φ_QEVF Valuation Dashboard
This notebook visualizes entanglement yield against revenue flux using the Φ_QEVF simulation model.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

from models.qevf_simulation import default_parameters, entanglement_to_revenue

data_path = Path("data/valuation/market_projection_2025-2035.csv")
market = pd.read_csv(data_path)
params = default_parameters()
entanglement_profile = [1e6, 3e6, 1e7, 3e7, 1e8]
revenue_curve = entanglement_to_revenue(entanglement_profile, params, revenue_scale=2.8)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(entanglement_profile, [r for _, r in revenue_curve], marker="o")
axes[0].set_title("Entanglement Yield vs Revenue Flux")
axes[0].set_xlabel("Entanglement Yield (logical ebits/s)")
axes[0].set_ylabel("Revenue Flux (USD/hour)")
axes[0].grid(True, alpha=0.3)

axes[1].plot(market["year"], market["valuation_low_billion"], label="Low Estimate", marker="s")
axes[1].plot(market["year"], market["valuation_high_billion"], label="High Estimate", marker="^")
axes[1].set_title("Market Valuation Trajectory (2025-2035)")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Valuation (Billion USD)")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
